# importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

# Loading Data

```python
data = pd.read_csv("insurance.csv")
```

```python
data.head()
```

# Data Cleaning

```python
data.info()
```

```python
data.describe()
```

# Feature Engineering

```python
Male = pd.get_dummies(data['sex'], drop_first = True)
data = pd.concat([data, Male], axis=1)
data.head()
```

```python
Smoker = pd.get_dummies(data['smoker'], drop_first = True)
data = pd.concat([data, Smoker], axis =1)
data.head()
```

```python
data = data.rename(columns = {'yes':'Smoker'})
data.head()
```

```python
data['region'].unique()
```

```python
Region = pd.get_dummies(data['region'], drop_first = True)
data = pd.concat([data, Region], axis=1)
data.head()
```

# splitting the data

```python
x=data.drop("charges",axis=1)
y=data["charges"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.4)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
data.head()
```

# scaling the data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaled_x_train = scaler.fit_transform(x_train)
scaled_x_test = scaler.transform(x_test)
```

# Support Vector Regression Model

```python
from sklearn.svm import SVR

regressor =  SVR()
regressor.fit(scaled_x_train, y_train)
```

```python
predict = regressor.predict(scaled_x_test)
```
